```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
#Declare main page addresses
url = "https://www.omicsonline.org/immunology-microbiology-journals.php "
#Launch Chrome browser and loading main page
browser = webdriver.Chrome('/Users/yuriysydorchuk/Downloads/chromedriver 2')
browser.get(url)
time.sleep(10)
```

```python
soup_level1=BeautifulSoup(browser.page_source, 'lxml')#Load page source


#Declare list where we will save link to our journals and PDF files
list_of_link=[]
list_of_publ_art_link=[]

#Select and save to "list_of_link" all link to our journals we need
td = soup_level1.html.body.find_all('td', {'valign':'middle'})
for t in td:
    link = t.select('a')[0].get('href')
    list_of_link.append(link)
```

```python
#Find and save all link to PDF files in our journal
for link in list_of_link:
        browser.get(link)
        time.sleep(13)
        soup_level1=BeautifulSoup(browser.page_source, 'lxml')
        
        
        #Get link to all pdf version of journals from first type of web page "omicsonline.org"
        if 'omicsonline.org' in link:
            td = soup_level1.html.body.find_all('h6', {'class':'font-size-4'})
            for t in td:
                if t.select('a')[0].get('href').startswith('https://'):
                    browser.get(t.select('a')[0].get('href'))
                    time.sleep(1)
                    soup_level=BeautifulSoup(browser.page_source, 'lxml')
                    td1 = soup_level.html.body.find_all('a', {'class':'btn btn-outline-warning'})
                    for t1 in td1:
                        if t1.get('href').endswith(".pdf"):
                            list_of_publ_art_link.append(t1.get('href'))
                        
                
        #Get link to all pdf version of journals from second type of web page "scitechnol.com"        
        elif 'scitechnol.com' in link:
            td = soup_level1.html.body.find_all('a', {'class':'btn btn-xs'})
            for t in td:
                if '.pdf' in t.get('href'): 
                    list_of_publ_art_link.append(t.get('href'))
                    
                    
        #Get link to all pdf version of journals from third type of web page "rroij.com"            
        elif 'rroij.com' in link:
            td = soup_level1.html.body.find_all('h4', {'class':'media-heading'})
            for t in td:
                if '<a href' in str(t):
                    browser.get(t.select('a')[0].get('href'))
                    time.sleep(1)
                    soup_level1=BeautifulSoup(browser.page_source, 'lxml')
                    td1 = soup_level1.html.body.find_all('a', {'class':'btn btn-default text-warning'})
                    for t1 in td1:
                        if t1.get('href').endswith('.pdf'):
                            list_of_publ_art_link.append(t1.get('href'))
                            
                            
        #Get link to all pdf version of journals from fourth type of web page which type like "imedpub.com"
        else:
            td = soup_level1.html.body.find_all('a', {'class':'btn btn-default'})
            for t in td: 
                if '.pdf' in t.get('href'): 
                    list_of_publ_art_link.append(link + t.get('href'))
        
browser.quit()
```

```python
#Check how much link to PDF file which we need
for i, link in enumerate(list_of_publ_art_link):
    print str(i+1) +') ' + link
```

```python
#Change preference of our browser for automatically download PDF files
chromeOptions=webdriver.ChromeOptions()
prefs = {"plugins.always_open_pdf_externally": True}
chromeOptions.add_experimental_option("prefs", prefs)

browser = webdriver.Chrome('/Users/yuriysydorchuk/Downloads/chromedriver 2', chrome_options=chromeOptions)
browser.get(url)
time.sleep(35)


#Download all file for which we have PDF link
for n, link in enumerate(list_of_publ_art_link):
    time.sleep(20)
    browser.get(link)
    print n+1, ':', link   

browser.quit()
```
